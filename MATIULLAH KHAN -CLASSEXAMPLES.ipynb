```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz
irisdata = datasets.load_iris()
print(irisdata)
measures = pd.DataFrame(irisdata.data, columns=irisdata.feature_names)
species = pd.Series(irisdata.target, name="species")
print(measures)
print(species)
classifier = DecisionTreeClassifier(criterion="gini", random_state=23456, max_depth=3, min_samples_leaf=5)
classifier.fit(measures, species)
classifier.predict([[4.9, 3.7, 2.6, 0.3]])
dot_data = tree.export_graphviz(classifier, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("iris")
```

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import graphviz

irisdata = datasets.load_iris()

print(irisdata)

measures = pd.DataFrame(irisdata.data, columns=irisdata.feature_names)
species = pd.Series(irisdata.target, name="species")
print(measures)
print(species)

classifier = DecisionTreeClassifier(criterion="gini", random_state=23456, max_depth=3, min_samples_leaf=5)
classifier.fit(measures, species)

classifier.predict([[4.9, 3.7, 2.6, 0.3]])

dot_data = tree.export_graphviz(classifier, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("iris")
```

```python
# Loading the file and replacing it? with NAN
kd=pd.read_csv('kd.csv',na_values='?')
```

```python
# Filter the data and print basic statistics
affected=kd.loc[(kd['class']== 'ckd')|(kd['class']=='ckd\t')]
affected.describe()
```

```python
correlations = affected.corr()
fig=plt.figure(figsize=[14,14])
ax=fig.add_subplot(111)
cax=ax.matshow(correlations,vmin=-1,vmax=1)
fig.colorbar(cax)
ticks=np.arange(0,11,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(kd)
ax.set_yticklabels(kd)

print(correlations)
plt.show()
```
