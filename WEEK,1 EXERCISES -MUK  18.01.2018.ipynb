```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x = np.arange(1000)
x
```

```python
y=np.log10(x+1)
y
```

```python
plt.plot(x,y)
plt.title('Graph of y =log(x+1)')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.grid()
plt.show()
```
